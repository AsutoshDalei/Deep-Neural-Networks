```python
_,_=plt.subplots(figsize=(18,25))
sns.set_style('dark')
for i in range(df.shape[1]):
    name=df.columns[i]
    plt.subplot(df.shape[1],1,i+1)
    plt.title(name)
    sns.lineplot(data=df[name])
plt.show()
```

```python
df2 = df.resample('D').sum()

# H is for hours, D is for days, W is for weeks

_,_=plt.subplots(figsize=(18,25))
sns.set_style('dark')
for i in range(df.shape[1]):
    name=df.columns[i]
    plt.subplot(df2.shape[1],1,i+1)
    plt.title(name)
    sns.lineplot(data=df2[name])
plt.show()
```
